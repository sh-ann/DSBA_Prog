```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\Аня\\Desktop\\useful\\DSBA\Prog\\norm\\Projects\Heart Disease Dataset.csv')
```

```python
'''
This part is AGE analysis
'''
print('Mean value of age:', df['age'].mean())
print('Median of age:', df['age'].median())
print('Standard deviation of age:', df['age'].std())
fig = plt.figure(figsize=(50, 20), dpi= 80, facecolor='w', edgecolor='k')
plt.xlabel("Age")
df['age'].plot.hist()
```

```python
'''
This part is SERUM CHOLESTORAL analysis
'''
print('Mean value of amount of serum cholestoral:', df['chol'].mean())
print('Median of amount of serum cholestoral:', df['chol'].median())
print('Standard deviation of amount of serum cholestoral:', df['chol'].std())
fig = plt.figure(figsize=(50, 16), dpi= 80, facecolor='w', edgecolor='k')
plt.xlabel("Level of serum cholestoral")
df['chol'].plot.hist()
```

```python
'''
This part is MAX HEART RATE analysis
'''
print('Mean value of maximum heart rate achieved:', df['thalach'].mean())
print('Median of maximum heart rate achieved:', df['thalach'].median())
print('Standard deviation of maximum heart rate achieved:', df['thalach'].std())
fig = plt.figure(figsize=(50, 16), dpi= 80, facecolor='w', edgecolor='k')
plt.xlabel("Person's number")
plt.ylabel('Heart rate')
df['thalach'].plot.bar()
```

```python
'''
This part is RESTING BLOOD PRESSURE analysis
'''
print('Mean value of resting blood pressure:', df['trestbps'].mean())
print('Median of resting blood pressure:', df['trestbps'].median())
print('Standard deviation of resting blood pressure:', df['trestbps'].std())
fig = plt.figure(figsize=(50, 16), dpi= 80, facecolor='w', edgecolor='k')
plt.xlabel("Person's number")
plt.ylabel('Blood pressure')
df['trestbps'].plot.bar()
```

```python
'''
Diagram for comparing max rate and resting blood pressure
I decided to do it usind the line graph, because it will be clear what is happening
'''
fig = plt.figure(figsize=(70, 20), dpi= 80, facecolor='w', edgecolor='k')
df['thalach'].plot(color="blue", label = 'MAX RATE')
df['trestbps'].plot(color="green", label = 'PRESSURE')
plt.xlabel("Person's number")
plt.legend(fontsize=50)
```

```python
'''
Here I'll show the depandence between age and the blood pressure
Firstly, I'll count average amount of serum for people from one age
Secondly, put it on the graph
'''
dic = [] 
all_ages = set()
for i in range(100):
    dic.append([i ,0, 0]) # first = age, second = numper of people with this age, third = sum of all amounts serum
for i in range(303):
    dic[int(df['age'][i])][1] += 1
    dic[int(df['age'][i])][2] += int(df['trestbps'][i])
    all_ages.add(int(df['age'][i]))
final_data = []
for i in range(100):
    if (dic[i][0] in all_ages):
        final_data.append([dic[i][0], dic[i][2] / dic[i][1]])
df2 = pd.DataFrame(np.array(final_data),
                   columns=['age', 'blood pressure'])
print(df2)
df2.plot(x='age', y='blood pressure', kind='bar', figsize=(50, 10))
plt.legend(fontsize=30)
```

```python
'''
Here I'll show the depandence between age and max rate
Firstly, I'll count average amount of serum for people from one age
Secondly, put it on the graph
'''
dic = [] 
all_ages = set()
for i in range(100):
    dic.append([i ,0, 0]) # first = age, second = numper of people with this age, third = sum of all amounts serum
for i in range(303):
    dic[int(df['age'][i])][1] += 1
    dic[int(df['age'][i])][2] += int(df['thalach'][i])
    all_ages.add(int(df['age'][i]))
final_data = []
for i in range(100):
    if (dic[i][0] in all_ages):
        final_data.append([dic[i][0], dic[i][2] / dic[i][1]])
df3 = pd.DataFrame(np.array(final_data),
                   columns=['age', 'Max Rate'])
print(df3)
df3.plot(x='age', y='Max Rate', kind='line', figsize=(50, 10))
plt.legend(fontsize=40)
```
